```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords.words('english')
```

```python
r_test="Hi my name is harshit verma and i love to do artificial language"
t_test_new=[word for word in r_test.split() if word.lower() not in stopwords.words('english')]
print(t_test_new)
```

```python
def review_cleaning(review):
    Test_punc_removed = [char for char in review if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join=Test_punc_removed_join.lower()
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
dp_vect_count=set()
reviews=dp['Review']
reviews
```

```python
for i,review in reviews.iteritems():
    review=review.strip('\n')
    review=review_cleaning(review)
    #print(review)
    for word in review:
        if(True):
            dp_vect_count.add(str(word))
#print(dp_vect_count)
dp_vect_count_list=list(dp_vect_count)
dp_vect_count_list=sorted(dp_vect_count_list)
dicti={}
count=0
for word in dp_vect_count_list:
    dicti[word]=count
    count=count+1
print(dp_vect_count_list)   
#print(dicti)
```

```python
r=[]
c=[]
v=[]
for i,review in reviews.iteritems():
    rr=str(review)
    rr=rr.strip('\n')
    rr=review_cleaning(rr)
    temp_dp=dict(Counter(rr))
    #print(temp_dp)
    for word,count in temp_dp.items():
        if(i==2):
            print(word)
        ig=dicti.get(word)
        if(ig!=None):
            r.append(i)
            c.append(ig)
            v.append(count)

            
DP_= ((csr_matrix((v, (r, c)), shape=(3600000, len(dicti)))).toarray())            
```

```python
from sklearn.model_selection import train_test_split
y=dp['Label']
```

```python
X_train,X_test,y_train,y_test=train_test_split(DP_,y,test_size=0.33,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
dp_bayes=MultinomialNB()
dp_bayes.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test = dp_bayes.predict(X_test)
dp_matrix = confusion_matrix(y_test, y_predict_test)
sea.heatmap(dp_matrix, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```
